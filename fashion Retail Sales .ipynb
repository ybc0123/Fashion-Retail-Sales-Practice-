```python
import pandas as pd

df = pd.read_csv("Fashion_Retail_Sales.csv")
df.head()
```

```python
#步驟一 資料清理
#轉換日期格式
df['Date Purchase'] = pd.to_datetime(df['Date Purchase'])
df.dtypes
```

```python
#建立月份欄位
df['Month'] = df['Date Purchase'].dt.to_period('M')
df.head()
```

```python
#檢查缺失值
df.isnull().sum() 
```

```python
#刪除缺失值
df_cleaned = df.dropna(subset = ['Purchase Amount (USD)','Review Rating'])
df = df_cleaned
#確保金額為float
df['Purchase Amount (USD)'] = pd.to_numeric(df['Purchase Amount (USD)'], errors='coerce')
```

```python
#步驟二 各商品銷售額總額
import matplotlib.pyplot as plt

item_sales = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().sort_values(ascending = False)
item_sales.plot(kind = 'bar', figsize = (12,5), title = 'Total sales amount of each product')
plt.ylabel('lump sum(USD)')
plt.grid(True)
plt.show()
```

```python
#步驟三 每月銷售趨勢
monthly_sales = df.groupby('Month')['Purchase Amount (USD)'].sum()
monthly_sales.plot(kind = 'line', color = 'orange', marker = 'o', figsize = (12,5), title = 'monthly sales trends')
plt.ylabel('lump sum(USD)')
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

```python
#步驟四 付款方式分布圖
df['Payment Method'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%', figsize = (6,6), title ='Payment Method Distribution')
plt.ylabel('')
plt.show
```

```python
#步驟五 評價與金額的關聯分析
import seaborn as sns

plt.figure(figsize = (8,5))
sns.scatterplot(data = df, x = 'Review Rating', y = 'Purchase Amount (USD)', color = 'orange')
plt.title('Relationship between Review Rating & Purchase Amount')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Step 1: 讀取 CSV 檔案
df = pd.read_csv('Fashion_Retail_Sales.csv')

# 檢查前幾筆資料確認是否正確
print(df.head())

# Step 2: 建立 SQLite 資料庫連線（存在記憶體中）
engine = create_engine("sqlite://", echo=False)

# Step 3: 將資料寫入名為 "sales" 的資料表中
df.to_sql("sales", con=engine, index=False, if_exists='replace')

# Step 4: 使用 SQL 查詢總銷售額
query = """
SELECT SUM([Purchase Amount (USD)]) AS total_revenue
FROM sales
"""
result = pd.read_sql_query(query, con=engine)
print(result)
```

```python
query = """
Select SUM([Purchase Amount (USD)]) as Tunic_revenue from sales
where [Item Purchased] = 'Tunic'
"""
result = pd.read_sql_query(query, con=engine)
print(result)
```

```python
query = """
Select SUM([Purchase Amount (USD)]) as Jeans_revenue from sales
where [Item Purchased] = 'Jeans'
"""
result = pd.read_sql_query(query, con=engine)
print(result)
```
